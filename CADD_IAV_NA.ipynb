```python
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski, MolFromSmiles
```

```python
def lipsinki(smiles, verbose=False):
  molecule_data = []
  for achar in smiles:
    molecule = MolFromSmiles(achar)
    molecule_data.append(molecule)
  baseData = np.arange(1,1)
  i =0
  for achar in molecule_data:
    des_MW = Descriptors.MolWt(achar)
    des_MolLogP = Descriptors.MolLogP(achar)
    des_NumHDonors = Lipinski.NumHDonors(achar)
    des_NumHAcceptors = Lipinski.NumHAcceptors(achar)
    row = np.array([des_MW, des_MolLogP, des_NumHDonors, des_NumHAcceptors])
    if i == 0:
      baseData = row
    else:
      baseData = np.vstack([baseData, row])
    i=i+1
  col_names= ["MW", "MolLogP", "NumHDonors", "NumHAcceptors"]
  descriptors = pd.DataFrame(data=baseData, columns=col_names)
  return descriptors
```

```python
#convert nM values in  'standard_value'  to molar concentrations
pIC50 = []
for aval in standard_value_list:
    if aval > 100000000:
        aval = 100000000
    molar_value = aval*(10**-9)
    pIC50.append(-np.log10(molar_value))
print(pIC50)
def norm_value(input):
    norm = []
    for i in input:
        if i > 100000000:
            i = 100000000
        norm.append(i)
    return norm
norm_pIC50 = norm_value(pIC50)
```

```python
import pandas as pd
from scipy.stats import mannwhitneyu

def mannwhitneyu_test(descriptor, df, alpha=0.05):
    df_active = df[df.bioactivity_class == 'active']
    df_active = df_active[descriptor]

    df_inactive = df[df.bioactivity_class == 'inactive']
    df_inactive = df_inactive[descriptor]
    active_data = df[df['bioactivity_class'] == 'active'][descriptor]
    inactive_data = df[df['bioactivity_class'] == 'inactive'][descriptor]

    stat, p = mannwhitneyu(active_data, inactive_data)
    interpretation = 'Different distribution (reject H0)' if p <= alpha else 'Same distribution (fail to reject H0)'

    results = pd.DataFrame({
        'Descriptor': descriptor,
        'Statistics': stat,
        'p': p,
        'alpha': alpha,
        'Interpretation': interpretation
    }, index=[0])

    return results
```

```python
dfan = df_for_analysis.dropna(how='any')
dfan
```

```python
df_X = dfan.drop('norm_pIC50', axis=1)
df_X.shape
```

```python
df_Y = dfan['norm_pIC50']
df_Y.shape
```

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(0.8*(1-0.8)))
df_X = selection.fit_transform(df_X)
df_X.shape
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_X, df_Y, test_size=0.2, random_state=42)
```

```python
train_x.shape, test_x.shape
```

```python
train_y.shape, test_y.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
np.random.seed(100)
model = RandomForestRegressor(n_estimators=100)
model.fit(train_x,train_y)
r2 = model.score(test_x,test_y)
r2
```

```python
Y_pred = model.predict(test_x)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(color_codes=True)
sns.set_style("white")

# Combine test_y and Y_pred into a single DataFrame for regplot
import pandas as pd
df = pd.DataFrame({'Experimental pIC50': test_y, 'Predicted pIC50': Y_pred})

ax = sns.regplot(x='Experimental pIC50', y='Predicted pIC50', data=df, scatter_kws={'alpha':0.4})
ax.set_xlabel('Experimental pIC50', fontsize='large', fontweight='bold')
ax.set_ylabel('Predicted pIC50', fontsize='large', fontweight='bold')
ax.set_xlim(0, 12)
ax.set_ylim(0, 12)
ax.figure.set_size_inches(5, 5)
plt.show()
```
